```python
import pandas as pd
import requests, sys
import csv

# Grab the genes from the gene_domain_mappings.txt file
genes=[]
mapped_to_domains = []

with open('gene_domain_mappings.txt','r') as f:
    next(f) # Skip the headers in the file
    reader=csv.reader(f,delimiter='\t')
    for i in reader:
        genes.append(i[0])
        mapped_to_domains.append(i[1])
```

```python
cosmic_gsme = pd.read_table('CosmicGenomeScreensMutantExport.tsv')
cosmic_genes = set(cosmic_gsme['Gene name'])
select_cosmic = cosmic_gsme[cosmic_gsme['Gene name'].isin(genes)]
```

```python
# Gene name - The gene name for which the data has been curated in COSMIC
# Mutation Id - unique mutation identifier
# Mutation CDS - The change that has occurred in the nucleotide sequence
# Mutation AA - The change that has occurred in the peptide sequence
# Mutation Description - Type of mutation (substitution, deletion, insertion, complex, fusion, unknown etc.)
# Mutation genome position - The genomic cooridnates of the mutation 

relevant_headers = ['Gene name', 'Mutation ID', 'Mutation CDS', 'Mutation AA', 'Mutation Description', 'Mutation genome position']
relevant_headers = ['Gene name', 'Mutation genome position']
mutation_info = select_cosmic[relevant_headers]
```
